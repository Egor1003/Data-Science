Получение hard-skill'ов из данных о навыках

```python
import re

def get_snippet(x):
    result = []
    en = re.compile('[a-zA-Z]+[+#.0-9]*')
    word = en.findall(x)
    if word:
        for i in word:
            result.append(i)
    if result == []:
        return np.NaN
    else:
        return result

```
